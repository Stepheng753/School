### Problem 1

Write a code to produce the Mandelbrot set and reproduce Fig 4.10 (hint: suppose that if $|z|>2$ then the orbit diverges to infinity).

<img src="figs/fig4.10.png" width="400" />


```python
def plot_mandelbrot_set(a_range, b_range, z0=0, max_iters=50):
    plot_a_vals = []
    plot_b_vals = []
    for a in a_range:
        for b in b_range:
            z = z0
            for _ in range(max_iters):
                z = z**2 + complex(a, b)
                z_mag = z.real**2 + z.imag**2
                if z_mag > 4:
                    plot_a_vals.append(a)
                    plot_b_vals.append(b)
                    break

    plt.grid()
    plt.plot(plot_a_vals, plot_b_vals, 'k,')
    plt.xlabel('Real($c$)')
    plt.ylabel('Imag($c$)')
    plt.title('Mandelbrot Set')

num_pts_per_axis = int(2e3)
plot_mandelbrot_set(np.linspace(-1.5, 0.6, num_pts_per_axis), np.linspace(-1.0, 1.0, num_pts_per_axis))
```

### Problem 2

Write a code to produce Julia sets and reproduce Fig 4.11.(a)

<img src="figs/fig4.11.png" width="400" />

```python
def plot_julia_set(a_range, b_range, c, max_iters=50):
    plot_a_vals = []
    plot_b_vals = []
    for a in a_range:
        for b in b_range:
            z = complex(a,b)
            try:
                for _ in range(max_iters):
                    z = z**2 + c
                    z_mag = z.real**2 + z.imag**2
                    if np.isinf(z_mag) or np.isnan(z_mag):
                        raise Exception("Invalid Number")
            except (RuntimeError, Exception):
                plot_a_vals.append(a)
                plot_b_vals.append(b)

    plt.grid()
    plt.plot(plot_a_vals, plot_b_vals, 'k,')
    plt.xlabel('Real($z$)')
    plt.ylabel('Imag($z$)')
    plt.title('Julia Set')

num_pts_per_axis = int(2e3)
c = complex(-0.17, 0.78)
plot_julia_set(np.linspace(-1.5, 1.5, num_pts_per_axis), np.linspace(-1.5, 1.5, num_pts_per_axis), c)
```

### Problem T4.8

Show that the box-counting dimension of a disk (a circle together with its interior) is 2.

Assume that we have a circle of radius $r$.  If there are boxes with side length $\epsilon$, then we can say the the radius in boxes is $r/\epsilon$.  Calculating the area of the circle in boxes, we get $N(\epsilon) = A = \pi(r/\epsilon)^2 = \pi r^2/\epsilon^2$. Now we can calculate the box counting dimension to be:
$$
\text{boxdim}(S) = \lim_{\epsilon \rightarrow 0} \frac{\ln N(\epsilon)}{\ln 1 / \epsilon} = \lim_{\epsilon \rightarrow 0} \frac{\ln \pi r^2 \epsilon^{-2}}{\ln \epsilon^{-1}} =  \lim_{\epsilon \rightarrow 0} \frac{\ln \pi r^2 -2 \ln \epsilon}{- \ln \epsilon} = \frac{\infty}{\infty} = \text{By L'H } \lim_{\epsilon \rightarrow 0} \frac{-2/\epsilon}{-1 / \epsilon} = 2 
$$

### Problem T4.9

Find the box-counting dimension of the invariant set of the skinny baker map of Example 4.5.

$$
\text{boxdim}(S) = \lim_{\epsilon \rightarrow 0} \frac{\ln N(\epsilon)}{\ln 1 / \epsilon} = \lim_{\epsilon \rightarrow 0} \frac{\ln 2^{\epsilon}}{\ln (1/(1/3))^\epsilon} = \lim_{\epsilon \rightarrow 0} \frac{\epsilon \ln 2}{\epsilon \ln (3)} = \frac{\ln 2}{\ln 3}
$$
