**Problem 1**: 10 pts - Here is a cubic polynomial with three closely spaced real roots:

$$
p(x) = 816x^3 - 3835x^2 + 6000x - 3125
$$

* What are the exact roots of p?
* Plot $p(x)$ for $1.43\leq x \leq 1.71$.  Show the location of the three roots.  
* Starting with $x_{0}=1.5$, what does Newton's method do?
* Starting with $x_{0}=1$ and $x_{1}=2$, what does the secant method do?
* Starting with the interval $[1,2]$, what does bisection do?

```python
def Newtons_method(f,fp,x0,tol):
    x1 = x0 - f(x0)/fp(x0)
    xapprox = np.array([x0,x1])
    while np.abs(x1-x0) >= tol:
        x0 = x1
        x1 = x0 - f(x0)/fp(x0)
        xapprox = np.append(xapprox,x1) 
    print ("Newtons Method   : Your Root is: %1.15f"%x1)
    #return xapprox[xapprox.size - 1]

def secant_method_mod(f,x0,x1,tol):
    fx0 = f(x0)
    fx1 = f(x1)
    x2 = x1 - ( (fx1 * (x1 - x0)) / (fx1 - fx0) ) 
    xapprox = np.array([x0,x1])
    while np.abs (x2 - x1) >= tol:
        x0 = x1
        x1 = x2
        fx0 = fx1
        fx1 = f(x1)
        x2 = x1 - ( (fx1 * (x1 - x0)) / (fx1 - fx0) )
        xapprox = np.append(xapprox, x2)
    print ("Secant Method    : Your Root is: %1.15f"%x2)
    #return xapprox[xapprox.size - 1]

def bisec_method(a,b,f,tol):
    fa,fb = f(a),f(b)
    if ( fa * fb ) < 0:
        c = ( b + a ) / 2
        capproxs = np.array([c])
        fc = f(c)
        while np.abs(fc) >= tol:
            if (fa * fc) < 0:
                b = c
            else:
                fa = fc
                a = c
            c = ( a + b ) / 2
            fc = f(c)
            capproxs = np.append(capproxs,c) 
        print ("Bisection Method : Your root is: %1.15f" %c)
        #return capproxs[capproxs.size -1]
    elif (fa == 0):
        print ("Bisection Method : Your root is: %1.1f" %fa)
    elif (fb == 0):
        print ("Bisection Method : Your root is: %1.1f" %fb)
    else:
        print ("Cannot ensure existence of root.")
```

```python
p = lambda x: (816 * (x**3))  - (3835 * (x**2)) + (6000 * x) - 3125
pp = lambda x: (816 * 3 * (x**2))  - (3835 * 2 * x) + (6000)

xvals = np.linspace(1.43, 1.71, int(1e3))
zero = np.zeros(xvals.size)

plt.plot(xvals, p(xvals))
plt.plot(xvals, zero)


print ("Exact Roots : 25/16, 25/17, 5/3\n")
Newtons_method(p,pp,1.5,1e-3)
secant_method_mod(p, 1, 2, 1e-3)
bisec_method(1,2,p,1e-3)
```

**Problem 3**: 10 pts - Suppose that we interpolate over three points, say $x_{j-1}$, $x_{j}$, $x_{j+1}$, with data $f_{j-1}$, $f_{j}$, and $f_{j+1}$, where our nodes are equally spaced so that 

$$
\delta x = x_{j+1} - x_{j} = x_{j} - x_{j-1}.
$$

Thus we can write 

$$
x_{j+1} = x_{j} + \delta x, ~ x_{j-1} = x_{j} - \delta x,
$$

As we discussed in class, for an unknown function $f(x)$, we can generate an interpolatory second-order approximation $P_{2}(x)$, $f(x)\approx P_{2}(x)$, where

$$
P_{2}(x) = f_{j-1}\frac{(x-x_{j})(x-x_{j+1})}{(x_{j-1}-x_{j})(x_{j-1}-x_{j+1})} + f_{j}\frac{(x-x_{j-1})(x-x_{j+1})}{(x_{j}-x_{j-1})(x_{j}-x_{j+1})} + f_{j+1}\frac{(x-x_{j-1})(x-x_{j})}{(x_{j+1}-x_{j-1})(x_{j+1}-x_{j})}
$$

Show that 

* (4pts) $P_{2}(x) = \frac{1}{2\delta x^{2}}\left(f_{j-1}\tilde{x}(\tilde{x}-\delta x) - 2f_{j}(\tilde{x}-\delta x)(\tilde{x}+\delta x) + f_{j+1}\tilde{x}(\tilde{x}+\delta x)\right)$, $\tilde{x} = x-x_{j}$.

* (3pts) $P'_{2}(x_{j}) = \frac{1}{2\delta x}\left(f_{j+1} - f_{j-1} \right)$.  Letting $f'(x_{j})\approx P'_{2}(x_{j})$ is known as a _centered difference_ approximation to a derivative of a function.

* (3pts) $P''_{2}(x_{j}) = \frac{1}{\delta x^{2}}\left(f_{j+1} - 2f_{j} + f_{j-1}\right)$

\begin{align}
&P_{2}(x) = f_{j-1}\frac{(x-x_{j})(x-x_{j+1})}{(x_{j-1}-x_{j})(x_{j-1}-x_{j+1})} + f_{j}\frac{(x-x_{j-1})(x-x_{j+1})}{(x_{j}-x_{j-1})(x_{j}-x_{j+1})} + f_{j+1}\frac{(x-x_{j-1})(x-x_{j})}{(x_{j+1}-x_{j-1})(x_{j+1}-x_{j})}\\\\
&x-x_{j} = \tilde{x}\\
&x_{j+1} - x_{j} = \delta x \\
&x_{j} - x_{j-1} = \delta x \\
&x-x_{j+1} = (x-x_{j}) - (x_{j+1} - x_{j}) = \tilde{x} - \delta{x} \\
&x-x_{j-1} = (x-x_{j}) + (x_{j} - x_{j-1}) = \tilde{x} + \delta{x} \\
&(x_{j+1}-x_{j-1})(x_{j+1}-x_{j}) = 2\delta{x}^{2}\\
&(x_{j}-x_{j-1})(x_{j}-x_{j+1}) = \delta{x}^{2}\\
&(x_{j-1}-x_{j})(x_{j-1}-x_{j+1}) = 2\delta{x}^{2}\\
&P_{2}(x) = \frac{1}{2\delta x^{2}}\left(f_{j-1}\tilde{x}(\tilde{x}-\delta x) - 2f_{j}(\tilde{x}-\delta x)(\tilde{x}+\delta x) + f_{j+1}\tilde{x}(\tilde{x}+\delta x)\right)\\
\\\\
&P'_{2}{(x)} = \frac{1}{2\delta{x}^{2}}(f_{j-1}(2\tilde{x} - \delta{x})+f_{j+1}(2\tilde{x} + \delta{x}) - 2f_{j}(2\tilde{x}))\\
&\text{let } g(x) = \tilde{x} \text{, then } g(x_{j}) = 0 \text{, so}\\
&P'_{2}{(x_{j})} = \frac{1}{2\delta x}\left(f_{j+1} - f_{j-1} \right)
\\\\
&P''_{2}{(x)} = \frac{1}{2\delta{x}^{2}}(f_{j-1}(2)+f_{j+1}(2) - 2f_{j}(2))\\
&P''_{2}(x_{j}) = \frac{1}{\delta x^{2}}\left(f_{j+1} - 2f_{j} + f_{j-1}\right)
\end{align}

```python
f = lambda x: np.cos(x)
fp = lambda x: -np.sin(x)
#   ( np.log(np.abs( fp(x[j]) - apx[j] ) ) ) = ( np.log(np.abs(C)) + P* np.log(np.abs(dx)))

def error(f,fp, N):
    dx = 2*np.pi / N
    x = np.zeros(N+1)
    aFPx = np.zeros(N-1)
    error = np.zeros(N-1)
    logError = np.zeros(N-1)
    for j in np.arange(0, int(N) + 1):   # j = 0, ..., N    size: N+1
        x[j] = (2*np.pi*j) / N
    for j in np.arange(0 + 1 , int(N) + 1 - 1):  # j = 1, ..., N-1    size:N+1 - 2terms(1st and last): N-1
        aFPx[j - 1] = ( f(x[j+1]) - f(x[j-1]) ) / (2*dx)
        #aFPx[0] = (f_2) - (f_0) / 2dx
        #aFPx[N-2] = (f_N) - (f_N-2) / 2dx
        error[j-1] = fp(x[j]) - aFPx[j-1]
        logError[j-1] = np.log(np.abs(error[j-1]))
    return logError

N = 100
jvals = np.arange(0 + 1 , int(N) + 1 - 1)
plt.plot(jvals, error(f,fp,N))
        
        
def maxError(f,fp,Nvals):
    maxError = np.array([1])
    maxError = np.delete(maxError,0)
    for N in Nvals:
        maxError = np.append(maxError, np.max(error(f,fp,N)))
    return maxError
    
xvals = np.arange(0,3+97)
nvals = np.arange(100,102+1+97)
plt.plot(xvals, maxError(f,fp,nvals))
print ("P = %1.15f" %((maxError(f,fp,nvals)[100 - 1] -  maxError(f,fp,nvals)[0] )/ (100 - 0)))
```
